```python
import pandas as pd
```

```python
listaAux=[['Peru', '190.237.60.38', True, '2021-11-02 02:17:53', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '190.237.60.38', True, '2021-11-02 02:16:52', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '190.187.147.146', True, '2021-11-02 02:13:55', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '190.237.60.38', True, '2021-11-02 02:13:13', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '190.237.60.38', True, '2021-11-02 02:12:23', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '190.237.60.38', True, '2021-11-02 02:10:28', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '179.6.204.26', True, '2021-11-02 02:06:38', '2021-11-02', 'PostmanRuntime/7.28.4'], ['Peru', '179.6.204.26', True, '2021-11-02 02:05:18', '2021-11-02', 'PostmanRuntime/7.28.4']]
```

```python
def setStatus(x):
    status="ENABLED"
    if(not x):
        status="DISABLED"
    return status
```

```python
dfIni = pd.DataFrame(listaAux, columns =['countryName','ipClient','active','dateTime', 'date','browser'])
dfIni
```

```python
dfBrowsers=dfIni[["dateTime","browser","active","countryName"]]
dfBrowsers
```

```python
dfReq=dfIni.groupby(['ipClient']).count()
dfReq=pd.DataFrame(dfReq, columns =['active'])
dfReq.reset_index(level=0, inplace=True)
dfReq.rename(columns = {'active': 'nRequests'}, inplace = True)
dfReq
```

```python
dfDatesMax=dfIni[['ipClient','dateTime']]
dfDatesMax=dfDatesMax.groupby(['ipClient']).max()
dfDatesMax.reset_index(level=0, inplace=True)
dfDatesMax
```

```python
mergedDetails = pd.merge(left=dfDatesMax, right=dfBrowsers, left_on='dateTime', right_on='dateTime')
mergedDetails['active'] = mergedDetails['active'].apply(setStatus)
mergedDetails.rename(columns = {'active': 'status'}, inplace = True)
mergedDetails
```

```python
mergeComplete = pd.merge(left=mergedDetails,right=dfReq,left_on="ipClient",right_on="ipClient")
mergeComplete.rename(columns = {'dateTime': 'lastAccess'}, inplace = True)
mergeComplete
```
